```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
filename = '../data/olympic_athlete_events.csv'

df = pd.read_csv(filename,
                usecols=['Age', 'Height', 'Team', 'Year', 'Season', 'Sport', 'Medal'])

df = df.loc[df['Team'].isin(['Great Britain', 'France', 'United States', 'Switzerland', 'China', 'India'])]
df = df.loc[df['Year'] >= 1980]

df.head()
```

```python
# What was the average age of olympic participants?
# In which country do medalists appear to consistently be the youngest?

pd.pivot_table(df, index='Year', columns='Team', values='Age')
```

```python
# How many medals did each country get in each year?
pd.pivot_table(df.dropna(subset='Medal'), 
               index='Year', 
               columns='Team', 
               values='Medal',
               aggfunc='size')
```

```python
# How tall was the tallest athlete in each sport in each year?
pd.pivot_table(df, index='Sport', columns='Year', values='Height', aggfunc='max')
```

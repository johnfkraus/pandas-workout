You analyzed math scores. If you perform the same analysis on verbal SAT

```python
import pandas as pd
import numpy as np
```

```python
math_cols = ['Year', 'State.Code', 'Total.Math', 'Family Income.Less than 20k.Math',
'Family Income.Between 20-40k.Math', 'Family Income.Between 40-60k.Math',
'Family Income.Between 60-80k.Math', 'Family Income.Between 80-100k.Math',
'Family Income.More than 100k.Math']
math_cols
```

```python
verbal_cols = []
for item in math_cols:
    verbal_item = item.replace('Math', 'Verbal')
    verbal_cols.append(verbal_item)

verbal_cols
```

```python
math_name_dict = {'Family Income.Less than 20k.Math': 'income<20k',
            'Family Income.Between 20-40k.Math': '20k<income<40k', 
             'Family Income.Between 40-60k.Math': '40k<income<60k',
            'Family Income.Between 60-80k.Math':'60k<income<80k', 
             'Family Income.Between 80-100k.Math':'80k<income<100k',
            'Family Income.More than 100k.Math':'income>100k'}
math_name_dict
```

```python
verbal_name_dict = {}
for key, value in math_name_dict.items():
    verbal_key = key.replace('Math', 'Verbal')
    verbal_name_dict[verbal_key] = value

verbal_name_dict
```

```python
df_math = pd.read_csv('../data/sat-scores.csv', header=0, usecols=math_cols)
```

```python
df_verbal = pd.read_csv('../data/sat-scores.csv', header=0, usecols=verbal_cols)
df_verbal.head()
```

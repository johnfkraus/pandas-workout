## Beyond 2

Which five states have the greatest gap in SAT math scores between the richest
and poorest students?

```python
df.columns
```

```python
# df[['State.Code', 'income<20k', 'income>100k']]
df['score_diff'] = df['income>100k'] - df['income<20k']
```

```python
(
df[['State.Code', 'income<20k', 'income>100k', 'score_diff']]
    .groupby('State.Code').mean()
    .sort_values(by='score_diff', ascending=False)
    .head()
)
```

```python
(
df[['State.Code', 'score_diff']]
    .groupby('State.Code').mean()
    .sort_values(by='score_diff', ascending=False)
    .head()
)
```
